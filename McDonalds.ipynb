Salads are by far the best Category in the Menu but does it mean the best profitable Category? 

```python
values = df['Category'].value_counts().sort_values(ascending=False)
labels = values.index
clrs=["red"]
sns.set_style("white")
plt.figure(figsize = (15,5))
sns.barplot(y=values, x=labels, palette=clrs)
plt.xticks(rotation=50)
plt.xlabel('Category')
plt.ylabel('Calories')
plt.title('Calories vs Category McDonals Menu');

plt.show()
```
